```python
# Import necessary libraries
!pip install pandas
!pip install holidays
import pandas as pd
import holidays

# Define the file path from OneDrive
file_path = r"C:\Users\zsmat\Pandora\Carlee Hammond - Master Thesis\Datasets\GB Traffic Data (Retail Cube)"

# Load the Excel file into a Pandas DataFrame
traffic_df = pd.read_excel(file_path, engine="openpyxl")

# Show the first few rows of the DataFrame
print(traffic_df.head())

# Get summary statistics of the 'Visitors CY' column
print(traffic_df.describe())

# Check for missing values
print(traffic_df.isnull().sum())

# Define the file path for the store data
store_file_path = r"C:\\Users\\caham\\OneDrive - Pandora\\Desktop\\Thesis\\Master Thesis\\Datasets\\Store data.xlsx"

# Load the Excel file into a Pandas DataFrame for store data
store_df = pd.read_excel(store_file_path, engine="openpyxl")

# Show the first few rows of the store DataFrame
print(store_df.head())

# Perform a left join on the Store ID columns
StoreTraffic_df = traffic_df.merge(store_df, on='Store ID', how='left')

# Show the merged DataFrame
print(StoreTraffic_df.head())

# Function to find missing dates for each store
def find_missing_dates(df):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    
    # Create a list to store results
    missing_dates_list = []
    
    # Group by Store ID
    for store_id, group in df.groupby('Store ID'):
        # Create a full date range for each store based on the first and last available date
        full_date_range = pd.date_range(start=group['Date'].min(), end=group['Date'].max(), freq='D')
        
        # Find the missing dates
        missing_dates = full_date_range.difference(group['Date'])
        
        # Only add the store to the results if there are missing dates
        if len(missing_dates) > 0:
            missing_dates_list.append({
                'Store ID': store_id,
                'Missing Dates Count': len(missing_dates),
                'Missing Dates': missing_dates
            })
    
    # Convert the list of missing dates to a DataFrame
    missing_dates_df = pd.DataFrame(missing_dates_list)
    
    return missing_dates_df

# Find missing dates in the store traffic data
missing_dates_df = find_missing_dates(StoreTraffic_df)

# Show the missing dates DataFrame
print(missing_dates_df.head())

# Install holidays package and get public holidays
!pip install holidays

# Function to get holidays for a country
def get_holidays(country_code):
    # Get the holidays for the given country and years
    holiday_instance = holidays.CountryHoliday(country_code, years=[2022, 2023, 2024, 2025])
    
    # Return a list of holidays
    return list(holiday_instance)

# Example usage for Great Britain
holidays_in_gb = get_holidays("GB")
print(holidays_in_gb)
```
